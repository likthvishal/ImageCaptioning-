```
import os
import sys
from glob import glob
import pandas as pd
import urllib.request
```

```
train_path = '/content/gdrive/My Drive/Colab Notebooks/Image Captioning/Train_GCC-training.tsv'
valid_path = '/content/gdrive/My Drive/Colab Notebooks/Image Captioning/Validation_GCC-1.1.0-Validation.tsv'
```

```
train_df = pd.read_csv(train_path, sep='\t', names=['caption', 'image_url'])
```

```
train_urls_ = train_df.image_url
```

```
valid_df = pd.read_csv(valid_path, sep='\t', names=['caption', 'image_url'])
```

```
valid_urls_ = valid_df.image_url
```

```
train_dest_path = '/content/gdrive/My Drive/Colab Notebooks/Image Captioning/Training'
if not os.path.isdir(train_dest_path):
    os.mkdir(train_dest_path)

valid_dest_path = '/content/gdrive/My Drive/Colab Notebooks/Image Captioning/Validation'
if not os.path.isdir(valid_dest_path):
    os.mkdir(valid_dest_path)
```

```
def download(urls, dest_path):
    for i in range(len(urls)):
        if i == 166:
            continue
        file_name = 'image_'+str(i)
        try:
            path = dest_path+'/'+file_name
            if not os.path.exists(path):
                urllib.request.urlretrieve(urls[i], path)
                print('Dowloaded image %d' % i)
        except Exception as e:
            print(e)
```

```
download(train_urls_, train_dest_path)
```

```
download(valid_urls_, valid_dest_path)
```
